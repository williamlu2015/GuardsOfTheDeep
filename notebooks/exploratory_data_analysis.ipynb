# CharCount and WordCount

```python
counts = [0, 0, 0]
for char_count in df["charCount"]:
    if char_count == 0:
        counts[0] += 1
    elif char_count < 10000:
        counts[1] += 1
    else:
        assert char_count >= 10000
        counts[2] += 1

plt.pie(counts, labels=[
    "zero", "regular", "outlier (>= 10000)"
], autopct="%.2f%%")

plt.axis("equal")
plt.title("charCount")
```

```python
char_counts = [
    char_count for char_count in df["charCount"]
    if 0 < char_count < 10000
]
bins = list(range(0, 10000 + 1, 100))

plt.hist(char_counts, bins=bins)
plt.xlabel("charCount")
plt.ylabel("Frequency")
```

```python
counts = [0, 0, 0]
for word_count in df["wordCount"]:
    if word_count == 0:
        counts[0] += 1
    elif word_count < 2000:
        counts[1] += 1
    else:
        assert word_count >= 2000
        counts[2] += 1

plt.pie(counts, labels=[
    "zero", "regular", "outlier (>= 2000)"
], autopct="%.2f%%")

plt.axis("equal")
plt.title("wordCount")
```

```python
word_counts = [
    word_count for word_count in df["wordCount"]
    if 0 < word_count < 2000
]
bins = list(range(0, 2000 + 1, 20))

plt.hist(word_counts, bins=bins)
plt.xlabel("wordCount")
plt.ylabel("Frequency")
```

```python
plt.scatter(df["wordCount"], df["charCount"], s=3)
plt.xlabel("wordCount")
plt.ylabel("charCount")
```

```python
nonzero_indices, = df["wordCount"].nonzero()
char_counts = df["charCount"].iloc[nonzero_indices]
word_counts = df["wordCount"].iloc[nonzero_indices]

chars_per_word = char_counts / word_counts
chars_per_word.describe()
```

```python
counts = [0, 0, 0]
for value in chars_per_word:
    if value < 5:
        counts[0] += 1
    elif 5 <= value < 7:
        counts[1] += 1
    else:
        assert value >= 7
        counts[2] += 1

display("Low outlier (< 5): " + "{:.2%}".format(counts[0] / len(chars_per_word)))
display("Regular: " + "{:.2%}".format(counts[1] / len(chars_per_word)))
display("High outlier (>= 7): " + "{:.2%}".format(counts[2] / len(chars_per_word)))
```

```python
plt.hist(chars_per_word, bins=np.linspace(5, 7, 100 + 1))
plt.xlabel("Characters per word")
plt.ylabel("Frequency")
plt.xticks(np.linspace(5, 7, 10 + 1))
```

# Headline

```python
df["headline"].size - df["headline"].dropna().size
```

```python
headlines = df["headline"].dropna()

counts = [0, 0]
for headline in headlines:
    if len(headline) < 100:
        counts[0] += 1
    else:
        assert len(headline) >= 100
        counts[1] += 1

display("Regular: " + "{:.2%}".format(counts[0] / len(headlines)))
display("Outlier (>= 100): " + "{:.2%}".format(counts[1] / len(headlines)))
```

```python
lengths = headlines.apply(len)

plt.hist(lengths, np.linspace(0, 100, 100 + 1))
plt.xlabel("Headline length (# characters)")
plt.ylabel("Frequency")
```

```python
def num_words(string):
    return len(string.split())

lengths = headlines.apply(num_words)
lengths.describe()
```

```python
plt.hist(lengths, np.linspace(0, 45, 45 + 1))
plt.xlabel("Headline length (# words)")
plt.ylabel("Frequency")
```

```python
df_filtered = df[
    df["headline"].notna()
]

headline_length = df_filtered["headline"].apply(len)
body_length = df_filtered["charCount"]

plt.scatter(headline_length, body_length, s=3)
plt.xlabel("Headline length (# characters)")
plt.ylabel("Body length (# characters)")
```

# StarRating

```python
counts = [0, 0]
for star_rating in df["starRating"]:
    if pd.isna(star_rating):
        counts[1] += 1
    else:
        counts[0] += 1

plt.pie(counts, labels=[
    "regular", "NaN"
], autopct="%.2f%%")

plt.axis("equal")
plt.title("starRating")
```

```python
counts = [0, 0, 0, 0, 0]
for star_rating in df["starRating"].dropna():
    counts[int(star_rating) - 1] += 1

plt.pie(counts, labels=[
    "1", "2", "3", "4", "5"
], autopct="%.2f%%")

plt.axis("equal")
plt.title("starRating")
```
