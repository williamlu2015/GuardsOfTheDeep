```python
import matplotlib.pyplot as plt

from ipynb.fs.full.data_loader import load_data
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from wordcloud import WordCloud
```

```python
df = load_data()
df.dropna(subset=["starRating"], inplace=True)
```

# Vectorizer

# Word cloud

## Full dataset

```python
text = " ".join(df["bodyText"])
```

```python
factory = WordCloud(
    width=1920, height=1080, prefer_horizontal=1, min_font_size=36, max_words=200, stopwords=None,
    background_color="white", max_font_size=72, collocations=False, normalize_plurals=False
)

words = factory.process_text(text)
```

```python
word_cloud = factory.generate_from_frequencies(words)
word_cloud.to_image()
```
