```python
thresholds = np.linspace(0.0002, 0.002, 10)
min_df_diff_var = []
var_diff_min_df = []

for threshold in thresholds:
    vectorizer = CountVectorizer(min_df=threshold)
    vectorizer.fit(corpus)
    feature_names_min_df = vectorizer.get_feature_names()
    
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(corpus)
    selector = VarianceThreshold(threshold)
    selector.fit(X)
    feature_names = vectorizer.get_feature_names()
    support = selector.get_support(indices=True)
    feature_names_var = [
        feature_names[index] for index in support
    ]
    
    min_df_diff_var.append(len(
        set(feature_names_min_df)
        .difference(set(feature_names_var))
    ))
    
    var_diff_min_df.append(len(
        set(feature_names_var)
        .difference(set(feature_names_min_df))
    ))
    
    display(f"Finished processing threshold {threshold}")

plt.plot(thresholds, min_df_diff_var, label="len(min_df - VarianceThreshold)")
plt.plot(thresholds, var_diff_min_df, label="len(VarianceThreshold - min_df)")

plt.xlabel("Thresholds")
plt.ylabel("Number of discrepancies")
plt.title("headline")
plt.legend()

plt.xticks(thresholds)
plt.ticklabel_format(style="sci", scilimits=(0, 0), axis="x")
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)

tfidf_scores = X.sum(axis=0).tolist()[0]
feature_names = vectorizer.get_feature_names()

assert len(tfidf_scores) == len(feature_names)
pairs = list(zip(feature_names, tfidf_scores))
pairs.sort(key=lambda x: x[1], reverse=True)

pairs
```

```python
stop_words = set(stopwords.words("english"))
vocabulary = [
    feature_name for feature_name, tfidf_score in pairs
    if feature_name not in stop_words
]
vocabulary = vocabulary[:100]

vocabulary
```
